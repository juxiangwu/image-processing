```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
import skimage
import skimage.io
import imageio
import math
```

```python
src = imageio.imread('../datas/desktop.jpg')
gray = cv2.cvtColor(src,cv2.COLOR_RGB2GRAY)
```

```python
edge = cv2.Canny(gray,50,200,1)
```

```python
plt.imshow(edge,'gray')
```

```python
lsd = cv2.createLineSegmentDetector(cv2.LSD_REFINE_STD)
```

```python
lines,width,prec,nfa = lsd.detect(edge)
```

```python
dst = lsd.drawSegments(src,lines)
```

```python
plt.imshow(dst)
```
