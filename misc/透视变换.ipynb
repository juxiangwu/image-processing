```python
#
corners = np.float32([[0,0],[cols - 1,0],[0,rows - 1],[cols - 1,rows - 1]])
corners_trans = np.float32([[150,150],[771,0],[0,rows - 1],[650,rows - 1]])

transform_maxtrix = cv2.getPerspectiveTransform(corners,corners_trans)
print(transform_maxtrix)
dst = cv2.warpPerspective( src , transform_maxtrix , (cols , rows))
plt.imshow(cv2.cvtColor(dst,cv2.COLOR_BGR2RGB))
```

```python
dst = cv2.perspectiveTransform(src,transform_maxtrix)
plt.imshow(cv2.cvtColor(dst,cv2.COLOR_BGR2RGB))
```
